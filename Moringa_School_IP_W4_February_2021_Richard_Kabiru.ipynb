# Data cleaning

**Validity**

```
# Irreleavant data is not needed. Therefore, I will drop columns that i find unnecessary.
car_df.drop(["Displayed comment", "Scheduled at"], axis=1, inplace= True)
car_df.info()
```

**Accuracy**

```
# Summing up the values in the bluecar_counter, utilib_counter and utilib_1.4_counter coolumns and assigning the values in the cars column.
car_df['cars'] = car_df['bluecar_counter'] + car_df['utilib_counter'] + car_df['utilib_1.4_counter']
car_df.head(10)
```

**Completeness**

```
# Checking for null values in the dataframe.
# If any exist, they will be dropped or filled with mode/mean. Impute
print(car_df.isnull().sum())
```

**Consistency**

```
# checking for any duplicates
car_df.duplicated()
```

**Uniformity**

```
# remove spaces from column names and adding an underscore 
car_df.columns = car_df.columns.str.strip().str.lower().str.replace(' ', '_')
car_df.info()
```

**Accuracy**

```
# Summing up the values in the bluecar_counter, utilib_counter and utilib_1.4_counter coolumns and assigning the values in the cars column.
car_df['cars'] = car_df['bluecar_counter'] + car_df['utilib_counter'] + car_df['utilib_1.4_counter']
car_df.head(10)
```

**Uniformity**

```
# Merging year, month and date into one column
car_df['date'] = car_df['day'].map(str) + '-' + car_df['month'].map(str) + '-' + car_df['year'].map(str) 
print (car_df) 
car_df.info()
```

```
# changing the data type of the new column named date to datetime.
car_df['date']= pd.to_datetime(car_df['date'])
car_df.info()
```

```
# combining the values in the hour and minute columns to create a new column in the database.
car_df['time'] = car_df['hour'].map(str) + ':' + car_df['minute'].map(str)
car_df.info()
```

```
# Converting the time column to datetime
car_df['time'] = pd.to_datetime(car_df['time'], format='%H:%M').dt.time
car_df.info()
```

```
# Checking the database after the cleaning exercise
car_df.head()
```

# Analysis

```
# Identifying the most popular hour of the day for picking up a shared electric car (Bluecar) in the city of Paris over the month of April 2018.
# popular hour of picking a bluecar in paris in april 2018
# columns to use city, bluecar == 0 and find the mode of time
pop_hour_bluecar = car_df['time'].groupby([car_df['city'] == 'Paris', car_df['bluecar_counter'] == 6])
pop_hour_bluecar.head()
```

```
# popular hour of picking a bluecar in paris in april 2018
# columns to use city, bluecar == 0 and find the mode of time
city_col = car_df['city'] == 'Paris'
pop_hour = car_df['hour']
blue_car = car_df['bluecar_counter'] == 0
pop_hour = car_df[city_col & pop_hour & blue_car]
pop_hour.sort_values(by= 'bluecar_counter', ascending= False)
```

```
# most popular hour of returning a car
# The mode will give us the most repeated value in the hour column. By sorting in descending order we get the values ordered from the highest to the lowest.

car_hour = car_df['hour'].mode()
car_hour.sort_values(ascending = False)
list(car_hour)
```
